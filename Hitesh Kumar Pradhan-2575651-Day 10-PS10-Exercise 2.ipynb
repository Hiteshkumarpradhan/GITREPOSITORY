# What are the ten most common movie names of all time?

```python

top_movie_titles = titles['title'].value_counts().reset_index()
top_movie_titles.columns = ['Title', 'Count']
top_movie_titles = top_movie_titles.sort_values(by='Count', ascending=False)
top_ten_movie_titles = top_movie_titles.head(10)
print(top_ten_movie_titles)
```

# Which three years of the 1930s saw the most films released?

```python
films_1930s = titles[(titles['year'] >= 1930) & (titles['year'] <= 1939)]
film_counts_by_year = films_1930s['year'].value_counts().reset_index()
film_counts_by_year.columns = ['Year', 'Number of Films Released']
film_counts_by_year = film_counts_by_year.sort_values(by='Number of Films Released', ascending=False)
top_three_years = film_counts_by_year.head(3)
print(top_three_years)
```

# Plot the number of films that have been released each decade over the history of cinema

```python
import pandas as pd
import matplotlib.pyplot as plt
titles = pd.read_csv('titles.csv')
decade_counts = titles.groupby((titles['year'] // 10) * 10).size()
plt.figure(figsize=(10, 6))
decade_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Films Released Each Decade')
plt.xlabel('Decade')
plt.ylabel('Number of Films')
plt.xticks(rotation=45)
plt.show()
```

# Plot how many roles Mammootty has played in each year of his career

```python
import matplotlib.pyplot as plt
cast = pd.read_csv('cast.csv')
mammootty_roles = cast[cast['name'] == 'Mammootty']
roles_per_year = mammootty_roles.groupby('year').size()
plt.figure(figsize=(12, 6))
plt.plot(roles_per_year.index, roles_per_year.values, marker='o', linestyle='-', color='b')
plt.title('Number of Roles Played by Mammootty Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Roles')
plt.grid(True)
plt.show()
```

# Plot the n-values of the roles that Judi Dench has played over her career.


```python
import pandas as pd
import matplotlib.pyplot as plt
cast = pd.read_csv('cast.csv')
judi_dench_roles = cast[cast['name'] == 'Judi Dench']
judi_dench_roles_with_n = judi_dench_roles[judi_dench_roles['n'].notna()]
plt.figure(figsize=(10, 6))
plt.scatter(judi_dench_roles_with_n['year'], judi_dench_roles_with_n['n'], alpha=0.5)
plt.title('n-values of Roles Played by Judi Dench Over Her Career')
plt.xlabel('Year')
plt.ylabel('n-value')
plt.grid(True)
plt.show()
```

# How many leading (n=1) roles were available to actors, and how many to actresses, in the 1950s?

```python
import pandas as pd
cast = pd.read_csv('cast.csv')
leading_roles_actors_1950s = cast[(cast['n'] == 1) & (cast['year'] // 10 == 195)]
leading_roles_actresses_1950s = cast[(cast['n'] == 1) & (cast['year'] // 10 == 195) & (cast['type'] == 'actress')]
num_leading_roles_actors_1950s = len(leading_roles_actors_1950s)
num_leading_roles_actresses_1950s = len(leading_roles_actresses_1950s)
print("Number of leading roles (n=1) for actors in the 1950s:", num_leading_roles_actors_1950s)
print("Number of leading roles (n=1) for actresses in the 1950s:", num_leading_roles_actresses_1950s)
```
