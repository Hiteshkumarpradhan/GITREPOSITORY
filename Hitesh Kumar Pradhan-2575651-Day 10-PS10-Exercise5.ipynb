# Make a bar plot of the months in which movies with "Christmas" In their title tend to be released in the USA.

```python
import pandas as pd
import matplotlib.pyplot as plt
christmas_movies_usa = release_dates[(release_dates['title'].str.contains('Christmas', case=False)) & (release_dates['country'] == 'USA')]
christmas_movies_usa['month'] = christmas_movies_usa['date'].dt.month
month_counts = christmas_movies_usa['month'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
month_counts.plot(kind='bar', color='skyblue')
plt.title('Number of "Christmas" Movies Released Each Month in the USA')
plt.xlabel('Month')
plt.ylabel('Number of Movies')
plt.xticks(rotation=0)
plt.show()
```

# Make a bar plot of the day of the week on which movies with "Romance" In their title tend to be released in the USA.


```python
import pandas as pd
import matplotlib.pyplot as plt
romance_movies_usa = release_dates[(release_dates['title'].str.contains('Romance', case=False, na=False)) & (release_dates['country'] == 'USA')]
romance_movies_usa['day_of_week'] = romance_movies_usa['date'].dt.day_name()
day_counts = romance_movies_usa['day_of_week'].value_counts()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_counts = day_counts.reindex(day_order)
plt.figure(figsize=(10, 6))
day_counts.plot(kind='bar', color='lightblue')
plt.title('Day of the Week for "Romance" Movies Released in the USA')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()
```
